```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('../data/application_train.csv')
print(df.head(10))
```

```python
print("Dataset Shape (Rows, Columns):", df.shape)
print("Total Number of Columns:", df.shape[1])
```

```python
print("\nTarget Distribution:")
target_dist = df['TARGET'].value_counts(normalize=True) * 100
print(target_dist)
```

```python
plt.figure(figsize=(6,4))
sns.countplot(data=df, x='TARGET')
plt.title('Target Distribution (0 = No Default, 1 = Default)')
plt.show()
```

```python
print("\nTop 10 Columns with Most Missing Values:")
missing = df.isnull().sum().sort_values(ascending=False).head(10)
print(missing)
```

```python
# Basic Statistics
print("\nBasic Statistical Summary:")
print(df.describe().T.head(10))  # first 10 numericals
```

```python
# Save summary for later
df.describe().to_csv('../data/data_summary.csv')
print("\n✅ EDA Completed! Summary saved in data/data_summary.csv")
```
